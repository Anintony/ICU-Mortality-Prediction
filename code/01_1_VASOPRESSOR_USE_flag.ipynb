```python
input_mv = pd.read_csv("/Users/jingyichen/machinelearning/a-mimic/mimic-iii-clinical-database-1.4/INPUTEVENTS_MV.csv")
input_mv.head()
```

```python
# Load datasets
icustays = pd.read_csv("/Users/jingyichen/machinelearning/a-mimic/mimic-iii-clinical-database-1.4/ICUSTAYS.csv")
d_items = pd.read_csv("/Users/jingyichen/machinelearning/a-mimic/mimic-iii-clinical-database-1.4/D_ITEMS.csv")
```

```python
# List of key drug names to find in both datasets
drug_names = ['Norepinephrine', "Levophed", 'Epinephrine', "Epinephrine-k", 'Dopamine', 'Phenylephrine', 'Vasopressin']

# Filter the D_ITEMS dataframe to get ITEMIDs for these drug names
drug_items = d_items[d_items['LABEL'].isin(drug_names)]
drug_items
```

```python
vasopressor_cv = [30043, 30044, 30051, 30047, 30119] 
# filter the INPUTEVENTS_CV dataset for these ITEMIDs
vasopressor_cv_data = input_cv[input_cv['ITEMID'].isin(vasopressor_cv)]

vasopressor_cv_data.info()
```

```python
# Filter the vasopressor_cv_data to include only relevant columns
vasopressor_cv_data = vasopressor_cv_data[['SUBJECT_ID', 'HADM_ID', 'ICUSTAY_ID']].copy()
vasopressor_cv_data['VASOPRESSOR_USE'] = 1

vasopressor_cv_data.info()
vasopressor_cv_data.head()
```

```python
# Remove duplicates
vasopressor_cv_flag = vasopressor_cv_data.drop_duplicates(subset=['SUBJECT_ID', 'HADM_ID', 'ICUSTAY_ID'])

vasopressor_cv_flag.info()
```

```python
# filter the INPUTEVENTS_MV dataset for these ITEMIDs
vasopressor_mv = [221662, 221289, 222315, 221906, 221749]

vasopressor_mv_data = input_mv[input_mv['ITEMID'].isin(vasopressor_mv)]

# Add flag for vasopressor use
vasopressor_mv_data = vasopressor_mv_data[['SUBJECT_ID', 'HADM_ID', 'ICUSTAY_ID']].copy()
vasopressor_mv_data['VASOPRESSOR_USE'] = 1

# drop duplicates
vasopressor_mv_flag = vasopressor_mv_data.drop_duplicates(subset=['SUBJECT_ID', 'HADM_ID', 'ICUSTAY_ID'])

vasopressor_mv_flag.head()
```

```python
vasopressor_all = pd.concat([vasopressor_mv_flag, vasopressor_cv_flag], axis=0)

vasopressor_all.info()
```

```python
# Keep only the relevant columns from icustays
icustays_core = icustays[['SUBJECT_ID', 'HADM_ID', 'ICUSTAY_ID']].copy()

# Left join to add vasopressor use flag to icustays
icustays_with_vaso = icustays_core.merge(
    vasopressor_all,
    on=['SUBJECT_ID', 'HADM_ID', 'ICUSTAY_ID'],
    how='left'
)

# Fill NaN values in VASOPRESSOR_USE with 0
icustays_with_vaso['VASOPRESSOR_USE'] = icustays_with_vaso['VASOPRESSOR_USE'].fillna(0).astype(int)


icustays_with_vaso.head()

```

```python
icustays_with_vaso.to_csv("/Users/jingyichen/Desktop/vasopressor_flagged_icustays.csv", index=False)
```
