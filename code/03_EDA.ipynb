#### Multivariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.countplot(
    x='insurance', 
    hue='hospital_expired', 
    data=df, 
    palette='Set1'
)

plt.title('Hospital Expiration by Insurance Type')
plt.xlabel('Insurance Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

plt.legend(
    title='Hospital Expired (0 = No, 1 = Yes)',
    title_fontsize='medium',
    fontsize='small'
)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(
    x='marital_status', 
    hue='hospital_expired', 
    data=df, 
    palette='Set1'
)

plt.title('Hospital Mortality by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

plt.legend(
    title='Hospital Expired\n(0 = No, 1 = Yes)',
    title_fontsize='medium',
    fontsize='small'
)

plt.tight_layout()
plt.show()
```

```python
pivot_table = df.pivot_table(index='marital_status', 
                             columns='insurance', 
                             aggfunc='size', 
                             fill_value=0)

insurance_order = ['Government', 'Medicaid', 'Medicare', 'Private', 'Self Pay']
pivot_table = pivot_table.reindex(columns=insurance_order)

pivot_table.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Insurance Type by Marital Status', fontsize=14)
plt.xlabel('Marital Status', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Insurance')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='icu_age', y='los_icu', hue='hospital_expired', data=df, alpha=0.6)

plt.title('ICU Age vs ICU Length of Stay by Mortality')
plt.xlabel('ICU Age')
plt.ylabel('Length of Stay in ICU')
plt.legend(title='Hospital Expired\n(0 = No, 1 = Yes)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
